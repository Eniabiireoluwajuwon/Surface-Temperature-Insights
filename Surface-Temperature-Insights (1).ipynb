```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

GLOBAL TEMPERATURE

```python
# Define the file paths for three CSV files
file1 = "global_temps.csv"  # Global temperatures CSV file
file2 = "nh_temps.csv"      # Northern Hemisphere temperatures CSV file
file3 = "sh_temps.csv"      # Southern Hemisphere temperatures CSV file
```

```python
# Read and print the contents of the global temperatures CSV file into a DataFrame (df1)
df1 = pd.read_csv(file1)
print(df1)
```

```python
# Display the first 10 rows of the DataFrame df1
df1_head = df1.head(10)
print(df1_head)

# Display the last 10 rows of the DataFrame df1
df1_tail = df1.tail(10)
print(df1_tail)
```

```python
# Rename the column 'J-D' to 'JD' in the DataFrame df1
# The inplace=False parameter ensures that a new DataFrame is returned, leaving df1 unchanged
df1 = df1.rename(columns={'J-D': 'JD'}, inplace=False)

# Print the DataFrame df1 after renaming the column
print(df1)
```

```python
df1.shape
```

```python
# Select only the 'Year' and 'JD' columns from the DataFrame df1
df1 = df1[['Year', 'JD']]
# Rename the columns to 'DS' and 'JD'
df1.columns = ['DS', 'JD']
# Set the 'DS' column as the index of the DataFrame
df1.set_index('DS', inplace=True)
# Remove rows with missing values (NaN) along the rows (axis=0)
df1.dropna(axis=0, inplace=True)

# Print the modified DataFrame df1
print(df1)
```

```python
# Create a training set (train) by selecting all rows up to the 12th-to-last row in the DataFrame df1
train = df1[:-12]

# Print the last few rows of the training set (train)
print(train.tail())
```

```python
# Set up a figure with a specific size (18 inches wide and 8 inches tall)
plt.figure(figsize=(18, 8))

# Plot the 'JD' values from the 'train' DataFrame against the index (assumed to be 'Year')
# with blue color and label 'Data'
plt.plot(train.index, train.JD, color="b", label='Data')

# Set the y-axis label
plt.ylabel('Change in Surface Temperature', fontsize=16)

# Set the x-axis label
plt.xlabel('Year', fontsize=16)

# Set the title of the plot
plt.title('Winter Global Surface Temperature Change Yearly Time Series', fontsize=20)

# Display the plot
plt.show()

```

```python
# Assuming 'train' is your DataFrame with columns 'index' and 'JD'
plt.figure(figsize=(18, 8))

# Scatter plot with regression line
sns.regplot(x=train.index, y=train.JD, color="b", scatter_kws={'s': 10}, line_kws={'color': 'red'})

# Set the y-axis label for the plot
plt.ylabel('Change in Surface Temperature', fontsize=16)

# Set the x-axis label for the plot
plt.xlabel('Year', fontsize=16)

# Set the title of the plot
plt.title('Global Surface Temperature Change Yearly Time Series with Regression Line', fontsize=20)

# Display the plot
plt.show()
```

```python

# Assuming 'train' is your DataFrame with columns 'index' and 'JD'
# For this example, let's assume you want to check the correlation between 'index' and ''

# Set up a figure with a specific size (10 inches wide and 6 inches tall)
plt.figure(figsize=(10, 6))

# Create a scatter plot using seaborn, with 'train.index' on the x-axis and 'train.JD' on the y-axis
sns.scatterplot(x=train.index, y=train.JD)

# Set the title of the scatter plot
plt.title('Scatter Plot of Correlation')

# Set the x-axis label
plt.xlabel('Year')

# Set the y-axis label
plt.ylabel('Change in Surface Temperature')

# Display the scatter plot
plt.show()

```

RELATIONSHIP BETWEEN WINTER IN THE NORTHERN AND SOUTHERN HEMISPHERE

```python
# Read the contents of the CSV file specified by 'file2' into a new DataFrame (df2)
df2 = pd.read_csv(file2)

# Read the contents of the CSV file specified by 'file3' into a new DataFrame (df3)
df3 = pd.read_csv(file3)

# Print the contents of DataFrame df2
print(df2)

# Print the contents of DataFrame df3
print(df3)
```

```python
# Select only the 'Year' and 'DJF' columns from the DataFrame df2
df2 = df2[['Year', 'DJF']]

# Rename the columns to 'DS' and 'DJF'
df2.columns = ['DS', 'DJF']

# Set the 'DS' column as the index of the DataFrame
df2.set_index('DS', inplace=True)

# Remove rows with missing values (NaN) along the rows (axis=0)
df2.dropna(axis=0, inplace=True)

# Print the modified DataFrame df2
print(df2)

# Select only the 'Year' and 'JJA' columns from the DataFrame df3
df3 = df3[['Year', 'JJA']]

# Rename the columns to 'DS' and 'JJA'
df3.columns = ['DS', 'JJA']

# Set the 'DS' column as the index of the DataFrame
df3.set_index('DS', inplace=True)

# Remove rows with missing values (NaN) along the rows (axis=0)
df3.dropna(axis=0, inplace=True)

# Print the modified DataFrame df3
print(df3)
```

```python
# Create a training set (train1) for df2 by selecting all rows up to the 12th-to-last row
train1 = df2[:-12]

# Print the first few rows of the training set (train1)
print(train1.head())

# Create a training set (train2) for df3 by selecting all rows up to the 12th-to-last row
train2 = df3[:-12]

# Print the first few rows of the training set (train2)
print(train2.head())
```

```python
# Set up a figure with a specific size (18 inches wide and 8 inches tall)
plt.figure(figsize=(18, 8))

# Plot the 'DJF' values from the 'train1' DataFrame against its index
# with blue color and label 'Northern Hemisphere'
plt.plot(train1.index, train1.DJF, color="b", label='Northern Hemisphere')

# Plot the 'JJA' values from the 'train2' DataFrame against its index
# with red color and label 'Southern Hemisphere'
plt.plot(train2.index, train2.JJA, color="r", label='Southern Hemisphere')

# Set the y-axis label
plt.ylabel('Change in Surface Temperature', fontsize=16)

# Set the x-axis label
plt.xlabel('Year', fontsize=16)

# Set the title of the plot
plt.title('WINTER SURFACE TEMPERATURE IN THE NORTHERN AND SOUTHERN HEMISPHERE', fontsize=20)

# Display the legend to differentiate between the two lines
plt.legend()

# Display the plot
plt.show()
```

```python
# Set up a figure with a specific size (18 inches wide and 8 inches tall)
plt.figure(figsize=(18, 8))

# Scatter plot with regression line for the Northern Hemisphere
sns.regplot(x=train1.index, y=train1.DJF, color="b", scatter_kws={'s': 10}, line_kws={'color': 'red'})

# Scatter plot with regression line for the Southern Hemisphere
sns.regplot(x=train2.index, y=train2.JJA, color="g", scatter_kws={'s': 10}, line_kws={'color': 'purple'})

# Set the y-axis label
plt.ylabel('Change in Surface Temperature', fontsize=16)

# Set the x-axis label
plt.xlabel('Year', fontsize=16)

# Set the title of the plot
plt.title('WINTER SURFACE TEMPERATURE IN THE NORTHERN AND SOUTHERN HEMISPHERE WITH REGRESSION LINES', fontsize=20)

# Display the plot
plt.show()
```

```python
# Set up a figure with a specific size (10 inches wide and 6 inches tall)
plt.figure(figsize=(10, 6))

# Scatter plot for the Northern Hemisphere with red color
sns.scatterplot(x=train1.index, y=train1.DJF, color="r", label='Northern Hemisphere')

# Scatter plot for the Southern Hemisphere with blue color
sns.scatterplot(x=train2.index, y=train2.JJA, color="b", label='Southern Hemisphere')

# Set the title of the scatter plot
plt.title('Scatter Plot of Correlation')

# Set the x-axis label
plt.xlabel('Year')

# Set the y-axis label
plt.ylabel('Change in Surface Temperature')

# Display the scatter plot
plt.show()
```
